```python
from unicorn_binance_websocket_api import BinanceWebSocketApiManager
from IPython.display import clear_output
import pandas as pd
import time
import sqlalchemy
from datetime import date
from datetime import datetime


bws = BinanceWebSocketApiManager(exchange="binance.com")
#bws = BinanceWebSocketApiManager(exchange="binance.com-testnet")
bws.create_stream(['trade'], ['btcusdt', 'bnbusdt', 'ethusdt'], output = "UnicornFy")
#bws.create_stream(['trade'], ['bnbusdt', 'ethusdt'], output = "UnicornFy")
engine_BTC = sqlalchemy.create_engine('sqlite:///BTCUSDTstream.db')
engine_ETH = sqlalchemy.create_engine('sqlite:///ETHUSDTstream.db')
engine_BNB = sqlalchemy.create_engine('sqlite:///BNBUSDTstream.db')

df = pd.DataFrame()

while True:
    oldest_data_from_stream_buffer = bws.pop_stream_data_from_stream_buffer()
    if oldest_data_from_stream_buffer:
        df = pd.DataFrame([oldest_data_from_stream_buffer])
        if df.columns[0] == 'result':
            print(df.columns)
        else:
            df.trade_time = pd.to_datetime(df.trade_time, unit='ms').item().strftime('%H:%M:%S.%f')
            df.event_time = pd.to_datetime(df.event_time, unit='ms').item().strftime('%H:%M:%S.%f')
            df = df.assign(Fecha_Actual=datetime.utcnow())
            df.Fecha_Actual = df.Fecha_Actual.item().strftime('%H:%M:%S.%f')
            df = df.loc[:,['Fecha_Actual', 'buyer_order_id', 'seller_order_id', 'trade_id', 'event_time', 'trade_time', 'symbol', 'price', 'quantity', 'is_market_maker']]
            df.columns = ["Fecha_Actual", "ID_pedido_comprador", "ID_pedido_vendedor", "ID_trade", "Tiempo_evento", "Tiempo_trade", "Simbolo", "Precio", "Cantidad", "Vendedor"]
            df['Precio'] = float(df['Precio'])
            df['Cantidad'] = float(df['Cantidad'])
            df = df.assign(Total_USD = float(df.Precio)*float(df.Cantidad))

            if df.Simbolo.item()  == 'ETHUSDT':
                df.to_sql(                                    
                    'ETHUSDT_6', 
                    engine_ETH, 
                    if_exists='append', 
                    index=False
                )
            if df.Simbolo.item()  == 'BNBUSDT':
                df.to_sql(                                    
                    'BNBUSDT_6', 
                    engine_BNB, 
                    if_exists='append', 
                    index=False
                    )
            if df.Simbolo.item()  == 'BTCUSDT':
                df.to_sql(                                    
                    'BTCUSDT_6', 
                    engine_BTC, 
                    if_exists='append', 
                    index=False
                )
    else: 
        print(df)
        clear_output(wait=True)
```
